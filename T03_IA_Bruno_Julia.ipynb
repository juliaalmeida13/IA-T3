```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import KFold

from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report


from sklearn.datasets import make_classification
#1 - Selecionar um conjunto de dados adequado para problemas de classificação (conjunto com classes nominais);
from sklearn.datasets import load_breast_cancer

from sklearn import tree
```

####**3** - Se for um conjunto de dados do pacote sklearn.datasets, carregar o conjunto com a função load_<nome_do_arquivo>;
####**4** - Separar o conjunto de dados em matriz de atributos (X) e vetor de classes (y).

```python
cancer = load_breast_cancer() 
X,y = load_breast_cancer(return_X_y= True)
```

####**6** - Implementar uma função que, dado um conjunto de dados, separe esse conjunto em 10 folds(10 subconjuntos).
####**7** - Dados os 10 folds, avaliar um algoritmo de indução de árvores de decisão usando a estratégia 10-fold cross-validation, ou seja, em cada uma das 10 iterações, treinar o algoritmo em 9 folds (folds de treinamento) e testar no fold restante (fold de teste), cada vez com um fold de teste diferente.
####**8** - Quando treinando o algoritmo de indução de árvore de decisão, usar a função fit(...) da classe DecisionTreeClassifier do módulo sklearn.tree, com o atributo criterion=’entropy’;
####**10** - Em cada um dos folds, fazer a classificação dos dados de teste usando a função predict(...) da classe DecisionTreeClassifier do módulo sklearn.tree;

```python
class_names = cancer.target_names
title = "Matriz de Confusão"
clf = tree.DecisionTreeClassifier(criterion='entropy')
kf = KFold(n_splits=10, shuffle= True)
predictFold = []
listClassReport = []
for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]
  
  clf = clf.fit(X_train, y_train)
  res = clf.predict(X_test)
  predictFold.append(clf)
  #11 - Em cada um dos folds, fazer a avaliação do modelo gerado usando os dados de teste e mostrando 
  #     os resultados das funções confusion_matrix(...) e classification_report(...) do modulo 
  #     sklearn.metrics;
  disp = plot_confusion_matrix(clf, X_test, y_test, display_labels = class_names,cmap = plt.cm.Blues)
  disp.ax_.set_title(title)
  plt.show()
  classReport = pd.DataFrame(classification_report(y_test, res, output_dict=True)).transpose()
  listClassReport.append(classReport)
  print("Relatório de avaliação do classificador: \n")
  print(classReport)
```

####**9** -  Escolha um dos folds para gerar a figura da árvore de decisão com a função plot_tree(...) do módulo sklearn.tree;

```python
fig = plt.figure(figsize=(20,20))
fig = tree.plot_tree(predictFold[3])
```

####12. Ao final das 10 execuções do 10-fold cross-validation, apresentar as médias e desvios padrões dos 10 resultados das medidas de avaliação da função classification_report(...) do modulo sklearn.metrics.

```python
media = listClassReport[0].copy()
desvio = listClassReport[0].copy()
for i in range(len(listClassReport[0])):
  for j in range(len(listClassReport[0].columns)):
    eachValors = []
    for report in listClassReport:
      eachValors.append(report.iloc[i][j])
    media.iloc[i][j] = np.mean(eachValors)
    desvio.iloc[i][j] = np.std(eachValors)
print("Média dos 10 relatórios de classificação")
print(media)
print("\nDesvio padrão dos 10 relatórios de classificação")
print(desvio)
```
